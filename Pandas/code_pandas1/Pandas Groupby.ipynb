```python
import pandas as pd

# Use 3 decimal places in output display
pd.set_option("display.precision", 3)

# Don't wrap repr(DataFrame) across additional lines
pd.set_option("display.expand_frame_repr", False)

# Set max rows displayed in output to 25
pd.set_option("display.max_rows", 25)
```

```python
raw_data=pd.read_csv("legislators-historical.csv")
```

```python
raw_data.columns
```

```python
df=raw_data[['last_name', 'first_name','birthday', 'gender', 'type', 'state','party']]
```

```python
df.head()
```

```python
df.describe()
```

```python
name_by_state=df.groupby("state")["first_name"].count()
```

```python
name_by_state
```

```python
name_gender_by_state=df.groupby(["state","gender"]).count()
```

```python
name_gender_by_state.head()
```

```python
name_gender_by_state["first_name"]
```

```python
by_state = df.groupby("state")
```
