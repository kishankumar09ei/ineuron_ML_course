```python
#defines a function of two scalar variables and returns a scalar result.
#The class vectorize can be used to “vectorize “this function so that

def addmultiply(a,b):
    if a>b:
        return a*b
    else:
        return a+b
```

```python
addmultiply(2,3)
```

```python
#The class vectorize can be used to “vectorize “this function so that
vec_addm=np.vectorize(addmultiply)
vec_addm([0,4,6,9],[1,6,5,7])
```

# Find the inverse of a matrix

```python
from scipy import linalg## linear algebra library
```

```python
from scipy import linalg ## linear algebra library
A = np.array([[1,3,5],[2,5,1],[2,3,8]])
print("Matrix: \n",A)
inv_mat = linalg.inv(A)
print("Inverse matrix:\n",inv_mat)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,200)
y = x*4
z = x**4
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,"b")
axes.set_xlabel("x")
axes.set_ylabel("y")
```

```python
axes.plot(x,y,"b")#look at this ..This wont work alone
```

```python
# we can also have more than two axes 

fig1=plt.figure()

axes1=fig1.add_axes([0.1,0.1,1,1])
axes2=fig1.add_axes([0.2,0.8,0.4,0.2])

axes1.plot(x,y,"b")
axes2.plot(x,z,"red")
```

```python
# Empty canvas of 1 by 2 subplots

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,8))
axes[0].plot(x,y,"b--")
axes[1].plot(x,z,"r--")
```
