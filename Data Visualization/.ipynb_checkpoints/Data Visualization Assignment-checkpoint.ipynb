**In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class**|

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
df=datasets.load_iris()
```

```python
X=df.data
y=df.target
```

**We will use PCA for reducing no of dimensions into 3**

```python
from sklearn.decomposition import PCA
import pandas as pd
pca=PCA(n_components=3)
principal_components=pca.fit_transform(X)
principal_df=pd.DataFrame(data=principal_components,columns=["pc1","pc2","pc3"])
```

```python
principal_df["species"]=y
```

```python
principal_df
```

```python
#now as we have our principal 3 components we can create our 3D plot

#creare a fig

fig=plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_zlabel('Principal Component 3', fontsize = 15)

ax.set_title('3 component PCA', fontsize = 20)

targets=[0,1,2]
species = ['setosa','versicolor','virginica']
colors = ['r','g','b']


for target,color in zip(targets,colors):
    species_index=principal_df["species"]==target
    ax.scatter(principal_df.loc[species_index,"pc1"],
               principal_df.loc[species_index,"pc2"],
               principal_df.loc[species_index,"pc3"],c=color)
ax.legend(species)
ax.grid()
    
    
    
```
